# Simple linear regression
# import relevant libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

# Load the data

```python
data=pd.read_csv(r"C:\Users\karan\OneDrive\Documents\real_estate_price_size.csv")
data
```

# Simple Linear Regression
# define dependent and independent variable

```python
y=data['price']
x1=data['size']
```

# explore the data

```python
plt.scatter(x1,y)
plt.xlabel('size',fontsize=20)
plt.ylabel('price',fontsize=20)
plt.show()
```

# regression itself

```python
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat=223.1787*x1+101900
fig=plt.plot(x1,yhat,lw=4,c='orange',label='regression line')
plt.xlabel('size',fontsize=20)
plt.ylabel('price',fontsize=20)
plt.show()

print("this regression line indicates that Size is a significant variable when predicting price of house in London")
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.score(x,y)         
```
