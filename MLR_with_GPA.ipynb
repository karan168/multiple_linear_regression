# sklearn Multiple linear regression and feature selection using standardization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv(r"F:\1.02.+Multiple+linear+regression.csv")
data
```

```python
data.describe()
```

# Create multiple linear regression
# Describe dependent and independent variables

```python
x = data[['SAT','Rand 1,2,3']]
y = data['GPA']  
```

# Regression itself

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

# R-squared

```python
reg.score(x,y)
```

# Adjusted R-squared

```python
r2 = reg.score(x,y)
n = x.shape[0]
p = x.shape[1]
adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```

```python
from sklearn.feature_selection import f_regression
```

```python
f_regression(x,y)
```

```python
p_value = f_regression(x,y)[1]
p_value
```

```python
p_value.round(3)      # 0 - SAT, 0.676 - Rand 1,2,3
```

# Creating a summary table

```python
reg_summary = pd.DataFrame(data = ['SAT','Rand 1,2,3'], columns = ['Features'])
reg_summary
```

```python
reg_summary['Coefficients'] = reg.coef_
reg_summary['p_value'] = p_value.round(3)
```

```python
reg_summary
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
x_scaled = scaler.transform(x)
```

```python
x_scaled
```

# Regression with scaled feature

```python
reg = LinearRegression()
```

```python
reg.fit(x_scaled,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Creating a summary table

```python
reg_summary = pd.DataFrame([['Bias'],['SAT'],['Rand 1,2,3']], columns = ['Features'])
reg_summary['Weight'] = reg.intercept_, reg.coef_[0], reg.coef_[1]
reg_summary
```

# 'SAT is a statistically significant and meaningful predictor of the dependent variable ( GPA). For every 1 unit increase in SAT, GPA increases by 0.172.
# Rand 1,2,3 is not significant and does not contribute to the predictive power of the model — it can probably be removed. For every 1 unit increase in this random variable, GPA slightly decreases by 0.007.
